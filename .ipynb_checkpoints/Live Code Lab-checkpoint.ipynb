```python
import pandas as pd
import numpy as np
import holoviews as hv
from holoviews import opts

hv.extension('matplotlib')
games = pd.read_csv('soccer-data.csv')
```

```python
vals = games[['home_team','away_team','home_score','away_score']].values
```

```python
def winner(vals):
    if vals[2] > vals[3]:
        return vals[0]
    elif vals[2] < vals[3]:
        return vals[1]
    else:
        return 'Tie'
```

```python
games['Winner'] = [winner(val) for val in vals]
```

```python
winners = games.Winner.value_counts().reset_index()
```

```python
winners.columns = ['Team','Winner']
```

```python
top10 = winners[winners.Team != 'Tie'].head(10)
```

```python
top10['Color'] = [
                   'red',
                   'yellow',
                   'orange',
                   'green',
                   'blue',
                   'gray',
                   'black',
                   'brown',
                   'purple',
                   'pink'
                   ]
```
